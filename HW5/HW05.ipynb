# Q1.寫一個函式，抓取Google Finance的股價資料

```python
import pandas_datareader.data as web
import matplotlib.pyplot as plt
from datetime import datetime
```

```python
import requests
import pandas as pd
def gprice(stock,startdate):
    params = {}
    params['q'] = stock
    params['startdate'] = startdate
    url = "https://finance.google.com/finance/historical?q={q}&startdate={startdate}&output=csv".format(**params)
    stockprice=pd.read_csv(url)
    print(stockprice)

gprice('AAPL',"2016-01-01")
```
